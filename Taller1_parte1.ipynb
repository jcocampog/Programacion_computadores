```python
#%% EJERCICIO 3
'''Este programa pide un número de 3 digitos y dice si existe el número 8'''

numero = int(input('Ingrese un número de tres cifras: '))
n = 8   #numero a encontrar en la cifra
if numero > 99 and numero < 1000:   #si el numero es positivo
    n3 = numero % 10    #n3 es el valor de unidad
    n2 = int((numero % 100)/10) #n2 es el valor de la decima
    n1 = int((numero % 1000)/100) #n1 es el valor de la centena
    if n1 == n or n2 == n or n3 == n:
        print('El número posee el dígito 8.')
    else:
        print('el número no posee el dígito 8.')
elif numero < -99 and numero > -1000: #si el numero es negativo
        d3 = -numero % 10   #d3 es el valor de unidad
        d2 = int((-numero % 100)/10)    #d2 es el valor de la decima
        d1 = int((-numero % 1000)/100)  #d1 es el valor de la centena
        if d1 == n or d2== n or d3 == n:
            print('El número posee el dígito 8.')
        else:
            print('El número no posee el dígito 8.')
else:
    print('El número no posee 3 dígitos.')
```

```python
#%% EJERCICIO 4
'''Este programa lee un número entero positivo y da la 
suma de sus números pares'''

numero = int(input('Ingrese un número entero: '))

pares=0

if numero < 0:
    print('El número no es un entero positivo')
else:
    while numero/10 > 0:    #mientras exista al menos una unidad
        n = numero % 10     #guardo el ultimo digito del numero
        if n % 2 == 0:      #si el modulo de ese digito es cero
            pares = pares + n   #agrego ese digito a 'pares'
        numero = int(numero/10)   #elimino ese ultimo digito, para seguir 
                                  #con el siguiente
    print('La suma de los pares del número ingresado es: ',pares)   
```

```python
#%% EJERCICIO 5
'''Este programa permite leer un número entero y determinar el número de cincos
   que están consecutivos en un número.'''

num_consecutivo = int(input('Ingrese el número a determinar consecutivo: '))
num = int(input('Escriba un número entero: '))                           
abs_num = abs(num) # Copia del valor absoluto de num para facilitar los cálculos

lista = []   #Lista que guarda los dígitos de num

while abs_num > 0:
    lista.append(abs_num %10)
    abs_num = int(abs_num/10)

lista.append(0)  #Se agrega un 0 a la lista de dígitos para que no falle el
                     #el indexado
cantidad = 0         #Contador de 5 consecutivos
for i in range(len(lista)):
    if lista[i] == num_consecutivo:
        if lista[i-1] == num_consecutivo or lista[i+1] == num_consecutivo:
            cantidad +=1

print(f'El número {num_consecutivo} se encuentra consecutivamente ',\
      f'{cantidad} veces.')
```

```python
#%% EJERCICIO 6
'''Este programa lee un número entero y dice si la suma de sus dígitos
   es un número de Fibonacci'''
def fib(n):                 #defino la función de fibonacci           
    fibonacci = []
    a,b = 1,1
    while a<=n:
        fibonacci.append(a)
        a,b = b,b+a
    return fibonacci

numero = int(input('Ingrese un número: '))
suma = 0

while numero/10 > 0:    #mientras exista al menos una unidad
    n = numero % 10     #guardo el ultimo digito del numero
    numero = int(numero/10) #elimino ese ultimo digito, para seguir 
                            #con el siguiente
    suma = suma + n     #voy sumando digito a digito
    
if suma in fib(suma):
    print('La suma de los dígitos del número ingresado es un número de \
          Fibonacci.')
else:
    print('La suma de los dígitos ingresado no es un número de Fibonnaci.')
```

```python
#%% EJERCICIO 7
''' Este programa lee un número entero positivo y determine si existen
en el número, dígitos repetidos.'''

numero = input('Ingrese un número: ')

repetido = []
unico = []

for x in numero:                #Empiezo a meter cada digito en 'unico'
	if x not in unico:
		unico.append(x)
	else:     # Si encuentra un número que esta 'unico', lo mete a 'repetido'
		if x not in repetido:
			repetido.append(x)
print('El valor que se repite es el: ',repetido)
```

```python
#%% EJERCICIO 8
'''Este programa itera a través de los primeros cien enteros positivos, 
   buscando los múltiplos de 3 e imprimiendo y almacenándolos en
   una lista hasta encontrar los primeros 15 de ellos. Una vez encontrados, 
   continua iterando en busca de los múltiplos de 4 y 
   almacenarlos en otra lista.'''
#%%Variables

rango_inferior = 1
rango_superior = 101
Limite_3 = 16
Inicio_4 = Limite_3-5
multiples_3=[3]  #Lista de los multiplos de 3
multiples_4=[]  #Lista de los multiplos de 4

#%% Función

def multiple(valor, multiple):

#%% Contidición de la función
    
    residuo = valor % multiple
    if residuo == 0:
        return True
    else:
        return False

#%%Multiplos de 3

for i in range(rango_inferior,rango_superior):#bucle entre los limites
    if multiple(i,3):   # si i es multiplo de 3
        multiples_3.append(i)   #lo agrego a la lista de los multiplos de 3
print('Los primeros 15 multiplos de 3 son: ',multiples_3[rango_inferior:\
                                                         Limite_3])

#%%Multiplos de 4

for j in range(rango_inferior,rango_superior): #bucle entre los limites
    if multiple(j,4):   # si i es multiplo de 4
        multiples_4.append(j) #lo agrego a la lista de los multiplos de 4
print('De los restantes números, los multiplos de 4 son: ',multiples_4\
      [Inicio_4:rango_superior]) #imprimo los multiplos de 4 entre el 
                                 #rango definido
```

```python
#%% EJERCICIO 9
'''Este programa permite leer las coordenadas (x1, y1, r1) y (x2, y2, r2) que 
corresponden al centro y al radio de dos círculos. Leer un punto de coordena-
das (a, b). Y determina si (a, b) está contenido: a) dentro del círculo 1; b) 
dentro del círculo 2; c) dentro de ambos círculos. Teniendo en cuenta que el 
círculo es el lugar geométrico de los puntos del plano cuya distancia a otro 
punto fijo, llamado centro, es menor o igual que una cantidad constante, 
llamada radio.'''

def distancia(x, y, a, b):
#la función perimite calcular la ditancia de dos puntos, fundamentada por el
#teorema de pitágoras
    dist = (((x - a) ** 2)+((y - b) ** 2)) ** (1 / 2)
    return dist

#%% Datos del círculo 1.
    
while True :
    try:
        coord_x_1 = int(input('Ingresar la coordenada en X del círculo 1: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intentelo de nuevo.')

while True :
    try:
        coord_y_1 = int(input('Ingresar la coordenada en Y del círculo 1: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intentelo de nuevo.')
        
#los ciclos while para las coordenadas del circulo 1 se usan para admitir uni-
#camente variables enteras, de manera que da la opción al usuario que si no lo
#hace, las ingrese nuevamente y de manera correcta.
        
while True :
    try:
        r_1 = int(input('Ingrese el radio del círculo 1: '))
        if r_1 >= 0 : break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')
#El while para el radio tiene como misma función acpetar unicamente enteros

#%% Datos del circulo 2.
    
while True :
    try:
        coord_x_2 = int(input('Ingresar la coordenada en X del círculo 2: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')

while True :
    try:
        coord_y_2 = int(input('Ingresar la coordenada en Y del círculo 2: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')
        
while True :
    try:
        r_2 = int(input('Por favor ingrese el radio del círculo 2: '))
        if r_2 >= 0 : break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')

#%% Ingresar las coordenadas del punto (a,b) a evaluar.
    
while True :
    try:
        a = int(input('Por favor ingrese la coordenada en X del punto: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')

while True :
    try:
        b = int(input('Por favor ingrese la coordenada en Y del punto: '))
        break
    except ValueError :
        print ('El valor ingresado no es correcto, intente de nuevo.')

#%% Ejecutar la función nombrada al inicio de código.
        
distancia_a_1 = distancia(coord_x_1, coord_y_1, a, b)
distancia_a_2 = distancia(coord_x_2, coord_y_2, a, b)

#%% Evaluar las condiciones de impresión.

if distancia_a_1 <= r_1 and distancia_a_2 <= r_2:
    print ('El punto', a,',', b, 'se encuentra dentro de ambos círculos.')
    
elif distancia_a_1 > r_1 and distancia_a_2 <= r_2:
    print ('El punto', a,',', b, 'se encuentra solo dentro del círculo 2.')
    
elif distancia_a_1 <= r_1 and distancia_a_2 > r_2:
    print ('El punto', a,',', b, 'se encuentra solo dentro del círculo 1.')
    
elif distancia_a_1 > r_1 and distancia_a_2 > r_2:
    print ('El punto', a,',', b, 'no se encuentra dentro de ningún círculo.')
```

```python
#%% EJERCICIO 11
''' Este programa dice de una cadena:
    -Cuántas vocales en mayúsculas hay
    -Cuántas letras con tilde
    -Cuántos dígitos tiene
    -Cuántos espacios tiene
    -Cuántas palabras reservadas se entraron '''
    
import keyword  
cadena = input('Ingrese una cadena de texto: ')

#%% Vocales en mayúsculas

contador = 0
voc_mayus=['A','E','I','O','U','Á','É','Í','Ó','Ú','Ü']

for x in cadena:
    if x in voc_mayus and cadena:
        contador += 1
print('La cantidad de vocales en mayusculas son: ',contador)

#%% Letras con tilde

contador2 = 0
voc_tildes = ['á','é','í','ó','ú','Á','É','Í','Ó','Ú']

for x in cadena:
    if x in voc_tildes and cadena:
        contador2 += 1
print('La cantidad de tildes son: ',contador2)

#%% Cantidad de dígitos

contador3 = 0

for x in cadena:
    if x.isdigit():
        contador3 += 1
print('La cantidad de dígitos que tiene la cadena son: ', contador3)

#%% Cantidad de espacio

contador4 = 0
espacio = [' ']

for x in cadena:
    if x in espacio and cadena:
        contador4 += 1
print('La cantidad de espacios son: ', contador4)

#%% Palabra reservadas

contador5 = 0
reservadas = keyword.kwlist
cad = cadena.split()

for x in cad:
    if x in reservadas and cad:
        contador5 += 1
print('La cantidad de palabras reservadas son: ',contador5)
```

```python
#%% EJERCICIO 13
'''Escribir un programa que lea una cadena de texto y la imprima como un 
   triángulo. Por ejemplo, si la cadena entrada es: "Es una locura odiar a 
   todas las rosas porque una te pinchó: renunciar a todos tus sueños porque 
   uno de ellos no se realizó - El Principito", el programa debe imprimir:
            E
           sun
          alocu
         raodiar
        atodaslas
       rosasporque
      unatepinchó:r
     enunciaratodost
    ussueñosporqueuno
   deellosnoserealizó-
      ElPrincipito
'''

cadena = input('Digite la cadena de texto: ')

# Se eliminan los espacios presentes en la cadena
cadena = cadena.replace(' ', '')

# Se determina el número de caracteres en la cadena
num_caract = len(cadena)

#%% Se determina el número de filas y el número máximo de caracteres por fila
temp = 0                       # Variable auxiliar 
num_filas = 1                  
while temp < num_caract:
    temp += 2 * num_filas - 1
    num_filas += 1

num_max = 2 * num_filas - 1    # Número máximo de caracteres en una fila

#%% Se imprime la cadena de texto 
aux_1 = 0        #Almacena la última posición en la que se imprime la fila i
aux_2 = 0        #Cantidad de letras que se imprimen en la fila i
contador = 0     # Verifica si luego de imprimir el triángulo aún quedan letras
                 # por imprimir

for i in range(1,num_filas):
    aux_2 += 2*i-1     
    print(cadena[aux_1:aux_2].center(num_max,' '))
    aux_1 = aux_2      
    contador += 1
    
if contador != num_caract:
    print(cadena[aux_2:].center(num_max,' '))
```

```python
#%% EJERCICIO 14
'''Este programa lee una cadena de texto y organiza alfabéticamente cada una
   de las letras que la componen, repitiendo cada una tantas veces como
   se encuentra.'''

frase = input('Ingrese una frase: ')    #Pide la frase
frase2 = frase.replace(" ", "")         #Quito los espacios
frase2 = sorted(frase2)                  #La organizo alfabeticamente
frase2 = ''.join(frase2)                #Junto la frase
print('El orden alfabetico de la palabra es: ',frase2)
```

```python
#%%% EJERCICIO 16
'''Leer una lista de números y decir si estan ordenados ascendentemente, 
   y luego pedir otro número y ordenarlo en la lista ascendente'''
    
n = int(input('¿Cuántos números desea ingresar? ')) #Pido el rango de la lista

lis_vacia=[]            #Creo una lista vacia

for i in range(n):
    a=int(input('Ingrese el número: '))   #Pido los numeros deacuerdo al rango
    lis_vacia.append(a)         #Agrego cada numero a la lista vacia
print('La lista de números ingresada es: ',lis_vacia)

lista = lis_vacia[:]        #Hago una copia de la lista
lis_vacia.sort()            #Ordeno la copia
if (lista == lis_vacia):    #Comparo la copia y el original
    print('La lista está correctamente ordenada.')
else:
    print('La lista no está ordenada correctamente.')

#%% Agregar otro número    
numero = int(input('Ingrese un número: '))

lis_vacia.append(numero) #le sumo el numero a la lista
lis_vacia.sort()        #Ordeno la nueva lista
print('La nueva lista ordenada es: ',lis_vacia)
```
